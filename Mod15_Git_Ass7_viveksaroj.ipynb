To perform Task 5 as shown in the image, will follow the **Git Flow workflow**, which typically includes these branches:

* `master` – production-ready code
* `develop` – integration branch for features
* `feature/*` – for new features
* `release/*` – for release preparation
* `hotfix/*` – for critical fixes on `master`

1. git flow init \
2. git flow feature start your-feature-name
3. git flow feature finish your-feature-name
#### This merges into 'develop' and deletes the feature branch
#### Then push changes:
4. git push origin develop
5. git flow release start v1.0
6. git flow release finish v1.0
7. git push origin master
8. git push origin develop
9. git flow hotfix start urgent-fix
#### Add `urgent.txt` file
10. git add urgent.txt
11. git commit -m "Add urgent.txt in hotfix"
12. git flow hotfix finish urgent-fix
13. git push origin master
14. git push origin develop





### Summary of Branch Flow:

* `feature` → `develop`
* `develop` → `release` → `master` (via release)
* `hotfix` → directly into `master` and `develop`
